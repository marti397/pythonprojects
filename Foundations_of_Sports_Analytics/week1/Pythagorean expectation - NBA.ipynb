```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt 
```

```python
nba = pd.read_csv('NBA_Games.csv')
nba
```

```python
print(nba.columns.tolist())
```

```python
nba18 = nba[nba['SEASON_ID'] == 22018.0]
nba18.describe()
```

```python
nba18 = nba[nba.SEASON_ID == 22018.0]
nba18.describe()
```

```python
nba18
```

```python
nba18.dropna(axis=0, inplace=True)
nba18
```

```python
nba18
```

```python
nba18['result'] = np.where(nba18['WL'] == 'W',1,0)
```

```python
nbateams18 = nba18.groupby("TEAM_ABBREVIATION")['result','PTS','PTSAGN'].sum().reset_index()
nbateams18
```

```python
nbateams18['wpc'] = nbateams18['result'] / 82
nbateams18['pyth'] = nbateams18['PTS']**2 / (nbateams18['PTS']**2 + nbateams18['PTSAGN']**2)
nbateams18
```

```python
sns.relplot(x="pyth", y="wpc", data = nbateams18)
```

```python
pyth_lm = smf.ols(formula = 'wpc ~ pyth', data=nbateams18).fit()
pyth_lm.summary()
```
