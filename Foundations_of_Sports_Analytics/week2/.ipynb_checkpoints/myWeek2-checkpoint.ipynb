```python
games = pd.read_csv('basketball_games.csv')
games.head(n=5)
```

```python
games.shape
```

```python
games.drop([0], inplace=True)
games.head(n=5)
```

```python
games = games[games['TEAM_ABBREVIATION'] != 'LVA']
```

```python
games.head()
```

```python
games = games[games['TEAM_ABBREVIATION'] != 'PHO']
```

```python
all_games = pd.merge(games, nba_teams, on=['TEAM_ID','TEAM_NAME'])
all_games.head(n=5)
```

```python
all_games.columns
```

```python
all_games.drop(columns=['ABBREVIATION'], inplace=True, errors='ignore')
```

```python
all_games.columns
```

```python
all_games.shape
```

```python
all_games.sort_values(by=['GAME_ID'])
```

```python
display(all_games)
```

```python
all_games.info()
```

```python
all_games.notnull()
```

```python
all_games[all_games.notnull()]
```

```python
all_games = all_games[pd.notnull(all_games['FG_PCT'])]
all_games.shape
```

```python
all_games = all_games.fillna(all_games.mean())
all_games.info()
```

```python
all_games['GM'] = all_games['FGM'] + all_games['FG3M'] + all_games['FTM']
```

```python
all_games['GA'] = all_games['FGA'] + all_games['FG3A'] + all_games['FTA']
```

```python
all_games['RESULT'] = np.where(all_games['PLUS_MINUS']>0, 'W' , "L")
```

```python
all_games.drop(['RESULT'], axis=1, inplace=True)
```

```python
all_games.sort_values(by=['GAME_ID', 'TEAM_NAME'])
```

```python
all_games.sort_values(by=['GAME_ID', 'WL'], inplace=True)
```

```python
all_games["POINTS_DIFF"] = all_games.groupby(['GAME_ID'])['PTS'].diff()
```

```python
all_games
```
